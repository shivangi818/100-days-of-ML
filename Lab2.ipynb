```python
x_train=np.array([1.0,2.0])
y_train=np.array([300.0,500.0])
print(f"x_train: {x_train}")
print(f"y_train: {y_train}")
```

```python
#m is the number of training examples
#x_train.shape returns a python tuple with an entry for each dimension
#x_train.shape[0] is the length of array
import numpy as np

# Assuming x_train is a list
x_train = np.array(x_train)

print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is: {m}")
```

```python
i=0
x_i=x_train[i]
y_i=y_train[i]
print(f"(x^({i}),y^({i}))=({x_i},{y_i})")
```

```python
#plotting the data
#The function arguments marker and c show the points as red crosses (the default is blue dots).
plt.scatter(x_train,y_train, marker='x' ,c='r')
#Set title
plt.title("Housing Prices")
#set the y-axix label
plt.ylabel("price in 1000s dollars")
plt.ylabel("sixe in 1000 sqft")
plt.show()
```

```python
#Model Function
#function that maps from x to y is represented as
# fw,b(x^i)=wx^i+b
#let w=100 and b=100
#The formula above is how you can represent straight lines - different values of and give you different straight lines on the plot.

w=100
b=100
print(f"w:{w}")
print(f"b:{b}")
```

```python
#Now, let's compute the value of 
#for your two data points. You can explicitly write this out for each data point as -

#for 
#x^0= f_wb = w * x[0] + b

#for 
#x^1= f_wb = w * x[1] + b
def compute_model_output(x,y,b):
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=w*x[i]+b
        
    return f_wb
```

```python
#Now let's call the compute_model_output function and plot the output..
tmp_f_wb= compute_model_output(x_train,w,b,)

#Plot our model prediction
plt.plot(x_train,tmp_f_wb,c='b',label='Our Prediction')

#Plot the data points
plt.scatter(x_train,y_train,marker='x',c='r',label='Actual Values')

#Ste title
plt.title("Housing Prices")
#Set the y_axis label
plt.ylabel('prices in 1000s dollars')
plt.xlabel('size in 1000 sqft')
plt.legend()
plt.show()
```
